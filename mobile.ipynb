```python
df['names'] = df['names'].fillna('NA notAvailable')
df['images_links'] = df['images_links'].fillna("Data not available")
df['stars'] = df['stars'].fillna(df['stars'].mode()[0])
df['rating&reviews'] = df['rating&reviews'].fillna(df['rating&reviews'].mode()[0])
df['price_details'] = df['price_details'].fillna("not availabel")
df['camara_info'] = df['camara_info'].fillna("not available")
df['display'] = df['display'].fillna("Not available")
df['battery'] = df['battery'].fillna("not available")
df['processor'] = df["processor"].fillna("not available")
df["warranty"] = df["warranty"].fillna("Not Available")
df["memory"] = df["memory"].fillna("not available")
```

Split the Review and Ratings value to new columns

```python
samsung.head()
ratings,reviews = [],[]
rating,review = [],[]
```

```python
for i in df['rating&reviews']:
    if '&' in i:
        x = i.split('&')
        ratings.append(x[0])
        reviews.append(x[-1])
    else:
        ratings.append("NA")
```

```python
def convert_to_str(i):
    y = i.split(",")
    z = ("".join(y))
    return(int(z))
```

```python
for i in ratings:
    y = i.split()
    rating.append(convert_to_str(y[0]))

for i in reviews:
    z=i.split()
    review.append(convert_to_str(y[0]))

```

```python
df['rating'] = rating
df['review'] = review
```

```python
df = df.drop(['rating&reviews'],axis=1,)
```

```python
df.sort_values(by="rating",ascending=False,inplace=True)
df.iloc[:10]
```

```python
price=[]
def price_correction(i):
    x = i.split("\n")
    y=x[0]
    z =y.split(",")
    p ="".join(z)
    price = slice(1,-1)
    return(int(p[price]))

for i in df['price_details']:

    try:
        price.append(price_correction(i))
    except:
        price.append(0)
```
